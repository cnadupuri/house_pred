```python
Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df["price"] >= lower) & (df["price"] <= upper)]
```

```python
x=df.drop('price',axis=1)
y=df['price']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python

ohe = OneHotEncoder(handle_unknown="ignore", sparse_output=False)

x_train_encoded = ohe.fit_transform(x_train[["city"]])
x_test_encoded = ohe.transform(x_test[["city"]])
```

```python
import pandas as pd

# Convert encoded city to DataFrame
city_train = pd.DataFrame(
    x_train_encoded,
    columns=ohe.get_feature_names_out(['city']),
    index=x_train.index
)

city_test = pd.DataFrame(
    x_test_encoded,
    columns=ohe.get_feature_names_out(['city']),
    index=x_test.index
)

# Remove original city column
x_train_num = x_train.drop('city', axis=1)
x_test_num = x_test.drop('city', axis=1)

# Combine numerical features + encoded city
x_train_final = pd.concat([x_train_num, city_train], axis=1)
x_test_final = pd.concat([x_test_num, city_test], axis=1)
```

```python
model.fit(x_train_encoded,y_train)
```

```python
y_pred=model.predict(x_test_encoded)
```

```python
from sklearn.metrics import accuracy_score,mean_absolute_error,mean_squared_error,r2_score,root_mean_squared_error
r2_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(random_state=42)

dt.fit(x_train_final, y_train)

y_pred = dt.predict(x_test_final)

print("Decision Tree R²:", r2_score(y_test, y_pred))
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gbr = GradientBoostingRegressor(random_state=42)

gbr.fit(x_train_final, y_train)

y_pred = gbr.predict(x_test_final)

print("Gradient Boosting R²:", r2_score(y_test, y_pred))
```

```python
from sklearn.ensemble import AdaBoostRegressor

ada = AdaBoostRegressor(random_state=42)

ada.fit(x_train_final, y_train)

y_pred = ada.predict(x_test_final)

print("AdaBoost R²:", r2_score(y_test, y_pred))
```

```python
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBRegressor

params = {
    "n_estimators": [100, 200, 300, 500],
    "max_depth": [3, 5, 7, 10],
    "learning_rate": [0.01, 0.05, 0.1],
    "subsample": [0.8, 1.0],
    "colsample_bytree": [0.8, 1.0]
}

xgb = XGBRegressor(random_state=42)

search = RandomizedSearchCV(
    xgb,
    param_distributions=params,
    n_iter=20,
    cv=5,
    scoring="r2",
    random_state=42,
    n_jobs=-1
)

search.fit(x_train_final, y_train)

print(search.best_params_)
print(search.best_score_)
```
